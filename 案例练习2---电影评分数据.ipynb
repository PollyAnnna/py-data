# 我们想要根据性别和年龄计算某部电影的平均得分

```python
data=pd.merge(pd.merge(users,ratings),movies)
data
```

```python
data.sort_values(by='user_id')
```

## 按性别计算每部电影的平均得分

### 我的方法：

```python
data1=data.groupby(['title','gender'])['rating'].mean()
data1
```

```python
data1.name='rating_avg'
data1.unstack(level=1)
```

### 书中方法：

```python
mean_ratings=data.pivot_table(index='title',columns='gender',values='rating',aggfunc='mean')
mean_ratings
```

## 过滤掉评分数据不够250条的电影

```python
title=data.groupby('title').size()
title
```

```python
title_new=title[title>=250]
title_new.index
```

```python
mean_ratings_title250=mean_ratings.loc[title_new.index]
mean_ratings_title250
```

### 女性观众最喜欢的电影:

```python
mean_ratings_title250.sort_values(by='F',ascending=False)
```

## 找出男性和女性观众分歧最大的电影:

### 我的方法：

```python
mean_ratings_title250.apply(lambda x:x.max()-x.min(),axis=1).sort_values(ascending=False)
```

### 书中方法：

```python
mean_ratings_title250['diff'] = mean_ratings_title250['M'] - mean_ratings_title250['F']
sorted_by_diff = mean_ratings_title250.sort_values(by='diff')
sorted_by_diff[:10]
```

## 只是想要找出分歧最大的电影（不考虑性别因素），则可以计算得分数据的方差或标准差：

```python
data.groupby('title')['rating'].std()[title_new.index].sort_values()
```
